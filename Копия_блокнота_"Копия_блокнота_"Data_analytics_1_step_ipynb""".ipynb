```python
import json
#from nltk import punctctu

filename = 'data.json'

data = {
    "string": "gshjkdf,:fdkjhsg",
    "consider_punctuation": False
}

with open(filename, 'w') as f:
    f.write(json.dumps(data, ensure_ascii=False, indent=4))

def palindrome(data: dict):
    if data["consider_punctuation"] == True:
      return data["string"] == data["string"][::-1]
    else:
      return data["string"].strip('.,;!:') == data["string"].strip('.,;!:')[::-1]


print(palindrome(data))
# В файле data.json дана строка из букв и знаков препинания.
# Проверить, является ли строка палиндромом.
# Также, в json указан параметр "consider_punctuation" - если False, то знаки препинания на "свойство палиндромности" не влияют, иначе влияют.
```
